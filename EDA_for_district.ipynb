```python
import pandas as pd
```

```python
dataframe = pd.read_csv("final.csv")
dataframe.loc[dataframe['District']=="hà nội  long biên", 'District'] = "long biên" 
dataframe.loc[dataframe['District']=="thị xã sơn tây", "District"] = "sơn tây"
dataframe.head()
```

```python
data = dataframe.groupby('District')['Price'].mean().round(0)
print(data)
for district in dataframe['District'].unique():
    dataframe.loc[dataframe[dataframe['District']==district].index,'Price_of_District'] = data[district]
```

```python
def get_center_latlong(df):
    # get the center of my map for plotting
    centerlat = (df['X'].max() + df['X'].min()) / 2
    centerlong = (df['Y'].max() + df['Y'].min()) / 2
    return centerlat, centerlong
```

```python
import folium
import branca.colormap as cm
center = get_center_latlong(dataframe)
feature = "Price_of_District"

colormap = cm.LinearColormap(colors=['green', 'yellow', 'yellow', 'yellow', 'red'], vmin=min(dataframe[feature]), vmax=max(dataframe[feature]))

m = folium.Map(location=center, zoom_start=10)

# go through each home in set, make circle, and add to map.
for i in range(len(dataframe)):
    folium.Circle(
        location=[dataframe.iloc[i]['X'], dataframe.iloc[i]['Y']],
        radius=10,
        fill=True,
        color=colormap(dataframe.iloc[i][feature]),
        fill_opacity=0.2
    ).add_to(m)
    """type_color = "blue"
    folium.Marker(
        location=[dataframe.iloc[i]['X'], dataframe.iloc[i]['Y']],
        popup=
                "Price: " + str(dataframe.Price[i]) + "<br>"
                + "Area: " + str(dataframe.Area[i]) + "<br>",
        icon=folium.Icon(color="%s" % type_color),
    ).add_to(m)"""


m.add_child(colormap)
```
